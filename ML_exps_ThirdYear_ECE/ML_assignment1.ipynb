```python
#CREATING A CSV FILE
import csv
# Sample data
data = [
    ["age", "income", "gender", "purchase_amount", "purchase_count"],
    [52, 57000, "M", 100, 2],
    [38, 70000, "F", 150, 4],
    [29, 45000, "F", 80, 1],
    [44, 80000, "M", 200, 2]
]
```

```python
#IMPORTING LIBRARIES AND LOADING THE DATA
import pandas as pd
# Load the dataset
data = pd.read_csv('dataset.csv')
```

```python
#Display the first few rows
print(data.head())

# Summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
#BIVARIATE ANALYSIS
# Scatter plot: Age vs. Income
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='income', data=data)
plt.title('Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
#Categorical Data Analysis
# Count plot: Gender distribution
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#UNIVARIATE ANALYSIS
import matplotlib.pyplot as plt
import seaborn as sns
# Histogram of ages
plt.figure(figsize=(8, 6))
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Outlier Detection
# Box plot: Income outliers
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.boxplot(data['income'])
plt.title('Income Box Plot')
plt.xlabel('Income')
plt.show()
```

```python
# Correlation heatmap
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data['gender_numeric'] = data['gender'].map({'F':0, 'M':1})

correlation_matrix = data.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
print('Quartile Q2 of sample=',np.quantile(s,0.5))
```

```python
print('Quartile Q3 of sample=',np.quantile(s,0.75))
```

```python
Q3=np.quantile(s,0.75)
Q1=np.quantile(s,0.25)
IQR=Q3-Q1
IQR
```

```python
UL=Q3+1.5*IQR
UL
```

```python
LL=Q1-1.5*IQR
LL
```

```python
outliers=[]
for i in s:
 if(i<LL or i>UL):
  outliers.append(i)
  print(outliers)
```

```python
#replace outliers by ul/ll
sam=np.array(s)
sam=np.where(sam>UL,UL,sam)
sam=np.where(sam<LL,LL,sam)
print('Original data samples',s)
print('After replacing outliers=',sam)
```

```python
#replace outliers by median
sam1=np.array(s)
sam1=np.where(sam1>UL,median,sam1)
sam1=np.where(sam1<LL,median,sam1)
print('Original data samples',s)
print('After replacing outliers=',sam1)
```
