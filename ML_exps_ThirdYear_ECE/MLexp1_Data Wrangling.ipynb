```python
import pandas as pd
df1 = pd.read_csv('E commerse details.csv')
df2 = pd.read_csv('emp_info.csv')
df3 = pd.read_csv('Employee Details.csv')
```

```python
dfs = {'file1': df1, 'file2': df2, 'file3': df3}
```

```python
for key, df in dfs.items():
    # Check for null values
    null_counts = df.isnull().sum()

    # Check if any null values exist in the DataFrame
    if null_counts.sum() > 0:
        print(f"DataFrame '{key}' has {null_counts.sum()} null values:")

        # Iterate through columns to print columns with null values
        for col, null_count in null_counts.items():
            if null_count > 0:
                print(f"  Column '{col}' has {null_count} null values")

                # Optionally, print indices where null values exist in the column
                if null_count > 0:
                    null_indices = df.index[df[col].isnull()].tolist()
                    print(f"    Null values at indices: {null_indices}")
    else:
        print(f"DataFrame '{key}' has no null values.")
```

```python
average_salary = df3['Salary'].mean()
print(average_salary)
```

```python
# Check if there are any null values in the 'salary' column
any_null_in_salary = df3['Salary'].isnull().any()
print(f"Are there any null values in 'salary' column? {any_null_in_salary}")
```

```python
# Calculate the average salary
average_salary = df3['Salary'].mean()

# Fill NaN values in 'Salary' column with average_salary and assign back to df3
df3['Salary'].fillna(average_salary, inplace=True)

# Step 3: Count NaN values in 'Salary' column (after fillna)
num_nulls = df3['Salary'].isnull().sum()
print(f"Number of null values in 'Salary' column after filling: {num_nulls}")

# Print the updated DataFrame to verify
print(df3)
```
