```python
# prompt: make an array of 10 elements, and find quartile1, Q2 and Q3 of this array, array elements are 12,15,34,60,64,75,87,88,94,98

import numpy as np
arr = np.array([12,15,34,60,64,75,87,88,94,98])
print("Q1 quantile of arr : ", np.quantile(arr, .25))
print("Q2 quantile of arr : ", np.quantile(arr, .50))
print("Q3 quantile of arr : ", np.quantile(arr, .75))
```
