```python
import numpy as np
import matplotlib.pyplot as plt

# Given dataset
class1 = np.array([[1, 2], [2, 3], [3, 3], [4, 5], [5, 5]])
class2 = np.array([[4, 2], [5, 0], [5, 2], [3, 2], [5, 3], [6, 3]])

# Combine the datasets
X = np.vstack((class1, class2))
y = np.array([0] * len(class1) + [1] * len(class2))  # Labels: 0 for class1, 1 for class2

# Step 1: Compute the mean vectors for each class
mean1 = np.mean(class1, axis=0)
mean2 = np.mean(class2, axis=0)

# Step 2: Compute the within-class scatter matrix
S1 = np.dot((class1 - mean1).T, (class1 - mean1))
S2 = np.dot((class2 - mean2).T, (class2 - mean2))
S_w = S1 + S2

# Step 3: Compute the between-class scatter matrix
mean_diff = (mean1 - mean2).reshape(-1, 1)
S_b = np.dot(mean_diff, mean_diff.T)

# Step 4: Compute the eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(np.linalg.inv(S_w).dot(S_b))

# Step 5: Select the eigenvector with the highest eigenvalue
idx = np.argmax(eigenvalues)
best_eigenvector = eigenvectors[:, idx]

# Step 6: Project the data onto the new space (1D)
X_lda = np.dot(X, best_eigenvector)

# Project each class separately
class1_lda = np.dot(class1, best_eigenvector)
class2_lda = np.dot(class2, best_eigenvector)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(class1_lda, np.zeros(len(class1_lda)), color='blue', label='Class 1')
plt.scatter(class2_lda, np.zeros(len(class2_lda)), color='red', label='Class 2')
plt.xlabel('LDA Component')
plt.title('LDA Projection of Class 1 and Class 2')
plt.legend()
plt.show()

# Displaying the transformation results
print("Eigenvector with highest eigenvalue:", best_eigenvector)
print("Transformed Data for Class 1 (1D):", class1_lda)
print("Transformed Data for Class 2 (1D):", class2_lda)
```
