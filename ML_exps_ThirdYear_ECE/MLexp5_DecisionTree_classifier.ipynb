# **Post Lab Questions:**

Handling missing data when constructing a decision tree is an important step to ensure the model remains robust and accurate. Here are several strategies you might consider:
1.   Imputation: Fill missing values with mean/median (for numerical) or mode (for categorical) values.

2.   Missing Indicator: Add a binary feature to indicate if data is missing.

3. Surrogate Splits: Use alternative splits for handling missing values during tree construction.
4. Mode Replacement: For categorical data, replace missing values with the most frequent category.
5.Exclude Instances: Remove rows with missing data, if the missingness is minimal.
6.Use Algorithms with Built-in Handling: Employ decision tree algorithms that manage missing data internally.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import tree
```

```python
from google.colab import files
uploaded = files.upload()

# Print the keys in the uploaded dictionary
print("Keys in uploaded:", uploaded.keys())
```

```python
df = pd.read_csv(io.BytesIO(uploaded['ObesityDataSet (1).csv']))
```

```python
# Explore the dataset
print("Head of the dataset:\n", df.head())
print("Tail of the dataset:\n", df.tail())
print("Info of the dataset:\n", df.info())
print("Describe the dataset:\n", df.describe())
print("Missing values:\n", df.isnull().sum())
```

```python
# Separate target and inputs
target = df["NObeyesdad"]
inputs = df.drop(["NObeyesdad"], axis=1)

# Encoding categorical variables
label_gender = LabelEncoder()
label_family_history = LabelEncoder()
label_favc = LabelEncoder()
label_fcvc = LabelEncoder()
label_ncp = LabelEncoder()
label_caec = LabelEncoder()
label_smoke = LabelEncoder()
label_scc = LabelEncoder()
label_faf = LabelEncoder()
label_tue = LabelEncoder()
label_calc = LabelEncoder()
label_mtrans = LabelEncoder()

# Apply encoding to the categorical columns
inputs["Gender"] = label_gender.fit_transform(inputs["Gender"])
inputs["family_history_with_overweight"] = label_family_history.fit_transform(inputs["family_history_with_overweight"])
inputs["FAVC"] = label_favc.fit_transform(inputs["FAVC"])
inputs["FCVC"] = label_fcvc.fit_transform(inputs["FCVC"])
inputs["NCP"] = label_ncp.fit_transform(inputs["NCP"])
inputs["CAEC"] = label_caec.fit_transform(inputs["CAEC"])
inputs["SMOKE"] = label_smoke.fit_transform(inputs["SMOKE"])
inputs["SCC"] = label_scc.fit_transform(inputs["SCC"])
inputs["FAF"] = label_faf.fit_transform(inputs["FAF"])
inputs["TUE"] = label_tue.fit_transform(inputs["TUE"])
inputs["CALC"] = label_calc.fit_transform(inputs["CALC"])
inputs["MTRANS"] = label_mtrans.fit_transform(inputs["MTRANS"])

# Display encoded inputs
print("Encoded inputs:\n", inputs.head())
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)

# Train the Decision Tree Classifier
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)

# Make predictions
y_predict = model.predict(x_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_predict)
print(f"The accuracy = {int(accuracy * 100)}%")

# Classification report
print("Classification Report:\n", classification_report(y_test, y_predict))
```

```python
plt.figure(figsize=(70, 30))  # Increase the figure size
plot_tree(model,
          feature_names=inputs.columns,
          class_names=model.classes_,
          filled=True,
          fontsize=20)  # Adjust font size if needed
plt.tight_layout()  # Automatically adjust subplot parameters
plt.show()
```
